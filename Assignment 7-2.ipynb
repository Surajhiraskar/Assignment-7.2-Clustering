```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data and perform EDA
airlines_data = pd.read_excel('/home/suraj/Downloads/Assignments/Assignment-7/EastWestAirlines.xlsx',sheet_name='data')
airlines_data
```

```python
airlines_data.isna().sum()
```

```python
airlines_data.dtypes
```

```python
# scaling the data
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(airlines_data.iloc[:,1:])
scaled_data
```

```python
airlines_data.columns
```

```python
scaled_df = pd.DataFrame(data=scaled_data,columns=['Balance', 'Qual_miles', 'cc1_miles', 'cc2_miles', 'cc3_miles',
       'Bonus_miles', 'Bonus_trans', 'Flight_miles_12mo', 'Flight_trans_12',
       'Days_since_enroll', 'Award?'])
scaled_df
```

```python
# Hierarchical clustering
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
dendogram = sch.dendrogram(sch.linkage(scaled_df))
```

```python
hc=AgglomerativeClustering(n_clusters=4, linkage='single')
hc
```

```python
y_hc = hc.fit_predict(scaled_df)
y_hc
```

```python
clusters = pd.DataFrame(data=y_hc, columns=['clusters'])
clusters.head()
```

```python
airlines_data_agglo = airlines_data.copy()
airlines_data_agglo['clusters'] = clusters
airlines_data_agglo.sort_values(by='clusters',ascending=False)
```

```python
airlines_data_agglo.groupby('clusters').agg(['mean'])
```

```python
# kmeans clustering
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)      # The elbow curve showing the gradual decrease at 3 in x axis
    
plt.plot(range(1,10),wcss)
plt.show() 
```

```python
# Building alogorithm
clusters_kmeans = KMeans(n_clusters=3,random_state=2)
clusters_kmeans.fit(scaled_df)
```

```python
clusters_kmeans.labels_
```

```python
clusters_k = pd.DataFrame(clusters_kmeans.labels_, columns= ['clusters'])
clusters_k.head()
```

```python
airlines_data_kmeans = airlines_data.copy()
airlines_data_kmeans['clusters'] = clusters_k
airlines_data_kmeans.sort_values(by='clusters',ascending=False)
```

```python
airlines_data_kmeans.groupby('clusters').agg(['mean'])
```

```python
# dbscan clustering
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(min_samples=6,eps=2)
dbscan.fit(scaled_df)
```

```python
dbscan.labels_
```

```python
clusters_d = pd.DataFrame(dbscan.labels_, columns= ['clusters'])
clusters_d.head()
```

```python
airlines_data_dbscan = airlines_data.copy()
airlines_data_dbscan['clusters'] = clusters_d
airlines_data_dbscan.sort_values(by='clusters',ascending=False)
```

```python
airlines_data_dbscan.groupby('clusters').agg(['mean'])
```
